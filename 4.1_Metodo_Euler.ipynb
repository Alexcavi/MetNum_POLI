```python
from pylab import *
import matplotlib.pyplot as plt

def euler(f, xo, yo, xf, n=10):
    y=[yo]
    x=[xo]
    h= (xf-xo)/n
    for i in range(1,n+1):
        yi=y[(i-1)]+h*f(x[(i-1)],y[(i-1)])
        xi=x[(i-1)]+h
        x.append(xi)
        y.append(yi)
    return x,y
```

```python
func = lambda t, y: 2 - 0.2*t
to = 0
yo = 5
n = 50
tf = 22
t,y = euler(func,to,yo,tf, n)
fig = plt.figure(figsize=(10, 10))
plt.plot(t,y, 'bx--', label='Euler')

# Si se tiene la curva soluciÃ³n
t_l = linspace(to , tf, 1000)
y_sol = lambda t: -0.1*t**2 + 2*t +5
plt.plot(t_l,y_sol(t_l), 'r', label='SoluciÃ³n')
# # # # # # # # # # # #

title("Curva de soluciÃ³n aproximada")
plt.show()

import pandas as pd
df= pd.DataFrame(zip(t,y),columns=["t","y"])
df
```
