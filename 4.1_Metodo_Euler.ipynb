```python
from pylab import *
import matplotlib.pyplot as plt

def euler(f,xo,yo,paso,xf):
    y=[yo]
    x=[xo]
    h=paso
    n=int((xf-xo)/h)
    for i in range(1,n+1):
        yi=y[(i-1)]+h*f(x[(i-1)],y[(i-1)])
        xi=x[(i-1)]+h
        x.append(xi)
        y.append(yi)
    return x,y
```

```python
func = lambda x, y: 2 - 0.2*x
posicion_x = 0
posicion_y = 5
paso = 0.5
t_final = 5
x,ye=euler(func,posicion_x,posicion_y,paso,t_final)
ye

plt.plot(x,ye, 'bx--', label='Euler')
legend(loc=4)
grid()
title("Curva de solución aproximada")

import pandas as pd
df=pd.DataFrame(zip(x,ye),columns=["Tiempo","Posición en X"])
df
```
